```
from sklearn.datasets import make_blobs #덩어리 데이터 
from scipy.spatial.distance import pdist #유클리드 거리
from scipy.cluster.hierarchy import linkage #계층적 군집화
from scipy.cluster.hierarchy import dendrogram #시각화 dendrogram 
import numpy as np #배열 사용
from matplotlib import pyplot as plt #차트 사용 

X,Y=make_blobs(n_samples=25,n_features=2,centers=3,cluster_std=1.5) #2차원, 표준편차가 1.5인 3개의 군집으로 25개의 샘플
colormap=np.array(['r','g','b']) #3가지의 색상으로 구분
plt.scatter(X[:,0],X[:,1],c=colormap[Y]) #산포도 표현
plt.title('Generated Data') #산포도 이름
plt.show()

Xdist=pdist(X,metric='euclidean') #거리행렬 계산
Z=linkage(Xdist,method='ward') #ward연결을 통한 계층적 군집화
Zd=dendrogram(Z) #dendrogram으로 표현
plt.title('Dendrogram') #그래프 이름
plt.show()
```
