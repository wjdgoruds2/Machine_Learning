```
from sklearn.datasets import make_moons #초승달모양 클러스터 생성
from sklearn.cluster import DBSCAN #DBSCAN
import numpy as np #배열 사용
from matplotlib import pyplot as plt #차트 사용 

X,Y=make_moons(n_samples=1000,noise=0.05) #표본수는 100개,잡음은 0.05

plt.title('Half moons')#산점도 이름
plt.scatter(X[:,0],X[:,1]) #산점도 표현
plt.show()

dbs=DBSCAN(eps=0.1) #떨어져야하는 기준 거리 0.01
Z=dbs.fit_predict(X) #군집화하기

colormap=np.array(['r','b']) #2가지 색으로 군집화
plt.scatter(X[:,0],X[:,1],c=colormap[Z]) #colormap을 통해 산점도 표현
plt.title('DBSCAN for half moons')#산점도 이름
plt.show()
```
