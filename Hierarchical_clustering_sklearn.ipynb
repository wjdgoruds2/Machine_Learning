```
from sklearn.datasets import make_blobs #덩어리 데이터 
from sklearn.cluster import AgglomerativeClustering #병합 군집 알고리즘
import numpy as np #배열 사용
from matplotlib import pyplot as plt #차트 사용 

X,Y=make_blobs(n_samples=2000,n_features=2,centers=8,cluster_std=2.0) #2차원, 표준편차가 2.0인 8개의 군집으로 2000개의 샘플
plt.scatter(X[:,0],X[:,1], s=4) #산포도 표현
plt.title('Generated Data') #산포도 이름
plt.show()

Z=AgglomerativeClustering(n_clusters=8,linkage='complete') #8개의 군집으로 complete(최대거리)연결로
P=Z.fit_predict(X) #군집화수행
colormap=np.array(['r','g','b','k','y','c','m','orange'])#8가지의 색을 통해 구분
plt.scatter(X[:,0],X[:,1],s=4,c=colormap[P]) #colormap을 통한 산포도 출력
plt.title('Clustering results') #산포도 이름
plt.show()
```
