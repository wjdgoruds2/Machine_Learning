```
import numpy as np #배열 사용
from matplotlib import pyplot as plt #차트 사용 

def gaussPlot(r,c,idx,mean,cov,title):
  x,y=np.random.multivariate_normal(mean,cov,400).T
  ax=plt.subplot(r,c,idx)
  ax.set_xlim(-10,10) #x축 -10~10까지 범위 설정
  ax.set_ylim(-10,10) #y축 -10~10까지 범위 설정
  ax.set_title(title)
  plt.plot(x,y,'.') #.으로 그래프 나타냄

mean=[0,0] #중심점
cov=[[1,0],[0,1]]
gaussPlot(2,2,1,mean,cov,'N([0,0],[[1,0],[0,1]])') #(0,0)을 중심점으로 [1 0][0 1]행렬 표현            

mean=[4,4] #중심점
cov=[[1,0],[0,1]]
gaussPlot(2,2,2,mean,cov,'N([4,4],[[1,0],[0,1]])') #(4,4)을 중심점으로 [1 0][0 1] 행렬 표현 

mean=[0,0] #중심점
cov=[[1,0],[0,5]]
gaussPlot(2,2,3,mean,cov,'N([0,0],[[1,0],[0,5]])') #(0,0)을 중심점으로 [1 0][0 5] 행렬 표현 

mean=[2,-3] #중심점
cov=[[1,2],[1,1]]
gaussPlot(2,2,4,mean,cov,'N([-2,-3],[[1,2],[1,1]])') #(-2,-3)을 중심점으로 [1 2][1 1] 행렬 표현 

plt.show()
```
