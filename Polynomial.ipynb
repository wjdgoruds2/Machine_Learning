```python
import numpy as np#벡터와 행렬 연산
import matplotlib.pyplot as plt#그래프 시각화 위함
from matplotlib.collections import LineCollection#삼중색상
from sklearn.linear_model import LinearRegression#선형 회귀
from sklearn.isotonic import IsotonicRegression#이소토닉 회귀
from sklearn.utils import check_random_state#랜덤 생성

n=100
x=np.arange(n)#100개 배열 생성
rs=check_random_state(0)
y=rs.randint(-50,50,size=(n,))+50.*np.log1p(np.arange(n))#-50~50사이 size형태의 배열 생성

ir=IsotonicRegression()#이소토닉 회귀
y_=ir.fit_transform(x,y)#fit하고 transform
lr=LinearRegression()#선형회귀
lr.fit(x[:,np.newaxis],y)#np.newaxis로 2차원 배열 형태 맞추고 데이터 변환

segements=[[[i,y[i]],[i,y_[i]]] for i in range(n)]#범위 설정
lc=LineCollection(segements,zorder=0)#삼중색으로 선표현 
lc.set_array(np.ones(len(y)))#len(y)배열을 1로 채워줌
lc.set_linewidths(np.full(n,0.5))#n배열을 0.5로 채워줌

fig=plt.figure()
plt.plot(x,y,'r.',markersize=10)#빨간색으로 10사이즈 산점도 출력
plt.plot(x,y_,'b.-',markersize=10)#파란색으로 10사이즈 산점도 출력
plt.plot(x,lr.predict(x[:,np.newaxis]),'g-')#노록색으로 예측값 그래프 출력
plt.gca().add_collection(lc)#객체 추출 좌표축 collection추가
plt.legend(('Data','Isotonic Fit','Linear Fit'),loc='lower right')#범례 설정
plt.title('Isotonic regression')#이름 설정
plt.show()#그래프 출력
```

```python
import numpy as np#벡터와 행렬 연산
import matplotlib.pyplot as plt#그래프 시각화 위함
from sklearn.linear_model import Ridge#리지회귀
from sklearn.preprocessing import PolynomialFeatures#다항 회귀
from sklearn.pipeline import make_pipeline#파이프라인 

def f(x):
  return x*np.sin(x)
x_plot=np.linspace(0,10,100)#0~10까지 100개의 요소
x=np.linspace(0,10,100)#0~10까지 100개의 요소
rng=np.random.RandomState(0)#난수 생성
rng.shuffle(x)#리스트 항목 섞기
x=np.sort(x[:20])#x배열0~19까지 정렬
y=f(x)#x*np.sin(x)값
X=x[:,np.newaxis]#np.newaxis로 2차원 배열 형태 맞춤
X_plot=x_plot[:,np.newaxis]#np.newaxis로 2차원 배열 형태 맞춤

colors=['teal','yellowgreen','gold']#색상설정
lw=2#선 굵기
plt.plot(x_plot,f(x_plot),color='cornflowerblue',linewidth=lw,label="ground truth")#cornflowerblue로 2굵기로 라벨은ground truth로 그래프 출력
plt.scatter(x,y,color='navy',s=30,marker='o',label="training points")#o마트로 라벨은training points로 navy색으로 산점도 

for count,degree in enumerate([3,4,5]):
  model=make_pipeline(PolynomialFeatures(degree),Ridge())#PolynomialFeatures(degree)와Ridge()값으로 파이프라인 연결
  model.fit(X,y)#X,y로 데이터 변환
  y_plot=model.predict(X_plot)#X_plot값으로 예측
  plt.plot(x_plot,y_plot,color=colors[count],linewidth=lw,label="degree %d"%degree)#설정한 색상으로 라벨은 각 degree에 맞게degree %d로 출력
plt.legend(loc="lower left")#범례 설정
plt.show()#그래프 출력
```
