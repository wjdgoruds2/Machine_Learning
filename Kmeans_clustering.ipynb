```
import numpy as np #배열 사용
from matplotlib import pyplot as plt #차트 사용 
from sklearn.datasets.samples_generator import make_blobs #덩어리 데이터 
from sklearn.cluster import KMeans #Kmeans군집화

X,Y=make_blobs(n_samples=300,centers=4,cluster_std=0.60,random_state=0) #표준편차가 0.60인 4개의 군집으로 300개의 샘플
plt.scatter(X[:,0],X[:,1], s=4) #산점도 표현,점의 크기 4
plt.title('Generated Data') #산점도 이름
plt.show()

kmeans=KMeans(n_clusters=4,init='k-means++',max_iter=300,n_init=10,random_state=0)#K-means++사용하여 4개의 군집으로 최대 300번 반복,초기중심점 반복회수 10번
#"K-means++" 는 하나의 무작위 중심점을 배정하고 두번째 중심점을 첫번째 중심점과 멀리 떨어진 곳에 배치, 그리고 다음 중심점을 1,2번째 중심점과 멀리 떨어지도록 배치하는 방식
pred_y=kmeans.fit_predict(X) #군집화 수행
colormap=np.array(['c','g','b','m'])#4가지의 색을 통해 구분
plt.scatter(X[:,0],X[:,1],s=4,c=colormap[pred_y]) #colormap을 통한 산점도 출력
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='red')  #산점도 표현,점의 크기 100,빨간색으로
plt.title('Clustering Results with Centers') #산포도 이름
plt.show()
```
